```python
gender = churn_df.groupby(['Gender' ])['Status'].value_counts(normalize = True).reset_index(name = 'rate')
gender
```

```python
gender['rate'] = gender['rate']*100
```

```python
gender
```

```python
fig = px.bar(gender, x = 'Gender', y = 'rate', color = 'Status', text_auto=True)
fig.show()
```

```python
def add_perc(rate):
  rate = round(rate, 0)
  return str(rate) + '%'
```

```python
gender['rate%'] = gender['rate'].apply(add_perc)
```

```python
gender
```

```python
fig = px.bar(gender, x = 'Gender', y = 'rate', color = 'Status', text='rate%')
fig.show()
```

```python
def age_group(Age):
  if Age in range(18, 50):
    return "Young Adult"
  elif Age in range(50, 60):
    return "Middle-aged Adult"
  else:
    return "Old Adult"  
```

```python
age_group(60)
```

```python
churn_df["Age_group"] = churn_df["Age"].apply(age_group)
```

```python
churn_df.head()
```

```python
age = churn_df.groupby(["Age_group", "Exited"])["Exited"].count().sort_values().reset_index(name = "count")
```
